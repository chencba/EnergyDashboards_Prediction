# Load UCD electricity data (2015-01-01 ~ 2022-07-19) and prediction data (2022-07-20 ~ 2022-10-17) and merge the two files for both campus and buildings

```python
import pandas as pd
```

```python
df_building = pd.read_csv("/content/UCD_electricity_buildings.csv")
df_building_prediction = pd.read_csv("/content/UCD_electricity_buildings_prediction.csv")
df_campus = pd.read_csv("/content/UCD_electricity_campus.csv")
df_campus_prediction = pd.read_csv("/content/UCD_electricity_campus_prediction.csv")
```

```python
df_building_prediction.rename(columns={df_building_prediction.columns[0]:"timestamp"},inplace=True)
df_campus_prediction.rename(columns={df_campus_prediction.columns[0]:"timestamp"},inplace=True)
```

```python
df_building = df_building[["timestamp","electricity_consumption","building"]]
df_building_new = pd.concat([df_building, df_building_prediction])
```

```python
df_campus = df_campus[["timestamp","electricity_consumption","campus"]]
df_campus_new = pd.concat([df_campus, df_campus_prediction])
```
